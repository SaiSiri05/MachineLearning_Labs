```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
np.random.seed(42)

X = np.linspace(-5, 5, 100)
y = 2 * X**2 - 3 * X + 5 + np.random.randn(100) * 2

# Initialize parameters
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()

# Hyperparameters
lr = 0.001
epochs = 5000

n = len(X)

losses = []

# Gradient Descent
for epoch in range(epochs):

    y_pred = a * X**2 + b * X + c

    loss = np.mean((y - y_pred) ** 2)
    losses.append(loss)

    # Gradients
    da = (-2/n) * np.sum((y - y_pred) * X**2)
    db = (-2/n) * np.sum((y - y_pred) * X)
    dc = (-2/n) * np.sum((y - y_pred))

    # Update
    a -= lr * da
    b -= lr * db
    c -= lr * dc

print("Learned parameters:")
print("a =", a)
print("b =", b)
print("c =", c)

# Plot
plt.scatter(X, y, label="Data")
plt.plot(X, a*X**2 + b*X + c, color='red', label="Fitted Curve")
plt.legend()
plt.show()
```

```python
learning_rates = [0.0001, 0.001, 0.01, 0.1]

for lr in learning_rates:

    x = 10
    losses = []

    for i in range(100):

        grad = 2 * x
        x = x - lr * grad

        losses.append(x**2)

    print(f"Learning Rate: {lr}, Final x: {x}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.linspace(-2, 2, 200)

y = X**5 - 4*X**4 + X**3 + 2*X**2 - X + 3

# Random initialization
coeffs = np.random.randn(6)

lr = 0.0001
epochs = 3000

for epoch in range(epochs):

    y_pred = (
        coeffs[0]*X**5 +
        coeffs[1]*X**4 +
        coeffs[2]*X**3 +
        coeffs[3]*X**2 +
        coeffs[4]*X +
        coeffs[5]
    )

    error = y - y_pred

    grads = []

    grads.append((-2/len(X)) * np.sum(error * X**5))
    grads.append((-2/len(X)) * np.sum(error * X**4))
    grads.append((-2/len(X)) * np.sum(error * X**3))
    grads.append((-2/len(X)) * np.sum(error * X**2))
    grads.append((-2/len(X)) * np.sum(error * X))
    grads.append((-2/len(X)) * np.sum(error))

    for i in range(6):
        coeffs[i] -= lr * grads[i]

print("Learned coefficients:")
print(coeffs)
```

```python
import numpy as np

# Dataset
X = np.random.randn(100)
y = 4 * X + 3

# Parameters
w = np.random.randn()
b = np.random.randn()

lr = 0.01
epochs = 100
batch_size = 10

for epoch in range(epochs):

    indices = np.random.permutation(len(X))

    X_shuffled = X[indices]
    y_shuffled = y[indices]

    for i in range(0, len(X), batch_size):

        X_batch = X_shuffled[i:i+batch_size]
        y_batch = y_shuffled[i:i+batch_size]

        y_pred = w * X_batch + b

        dw = (-2/len(X_batch)) * np.sum((y_batch - y_pred) * X_batch)
        db = (-2/len(X_batch)) * np.sum(y_batch - y_pred)

        w -= lr * dw
        b -= lr * db

print("w =", w)
print("b =", b)
```

```python
import numpy as np

x = 10

lr = 0.1
momentum = 0.9

velocity = 0

for i in range(50):

    grad = 2 * x

    velocity = momentum * velocity - lr * grad

    x = x + velocity

print("Final x =", x)
```
