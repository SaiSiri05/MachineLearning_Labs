```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

```python
data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))

X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24, 7))
plt.plot(X["make"], y)
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Car maker vs Price - Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(24, 7))
plt.bar(X["make"], y)
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Car maker vs Price - Bar Chart')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y, bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Car Prices - Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(y)
plt.ylabel('Price')
plt.title('Car Price Distribution - Box Plot')
plt.show()
```

```python
make_counts = X["make"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%')
plt.title('Car Make Distribution - Pie Chart')
plt.show()
```

```python
make_counts = X["make"].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(make_counts.index, make_counts.values)
plt.xlabel('Make')
plt.ylabel('Count')
plt.title('Number of Cars per Make')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 6))
sns.heatmap(X.select_dtypes(include=np.number).corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data=data[["price", "engine-size", "horsepower", "city-mpg"]])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(x=X["fuel-type"], y=y)
plt.title('Fuel Type vs Price (Swarm Plot)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=X["engine-size"], y=y)
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Regression Plot')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(x=X["fuel-type"], y=y)
plt.title('Fuel Type vs Price (Strip Plot)')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Box Plot
sns.boxplot(x=X["body-style"], y=y, ax=axes[0])
axes[0].set_title('Box Plot: Body Style vs Price')
axes[0].set_xlabel('Body Style')
axes[0].set_ylabel('Price')
axes[0].tick_params(axis='x', rotation=45)

# Violin Plot
sns.violinplot(x=X["body-style"], y=y, ax=axes[1])
axes[1].set_title('Violin Plot: Body Style vs Price')
axes[1].set_xlabel('Body Style')
axes[1].set_ylabel('Price')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Different Types of Charts & Graphs:

   1. Scatter Plot

            Use: Relationship between two numerical variables
            Example: Engine size vs Price

            Key points:

            Shows correlation (positive / negative / none)
            Helps detect patterns & clusters

   2. Line Plot

            Use: Trends over time or ordered data

            Key points:

            Best for time-series (year, date)
            Shows increase/decrease clearly
   3. Bar Plot

            Use: Compare values across categories

            Key points:

            In Seaborn → shows average (mean) by default
            Good for categorical vs numerical

   4. Count Plot

            Use: Frequency of categories

            Key points:

            Only categorical variable needed
            Equivalent to value_counts()

   5. Histogram

          Use: Distribution of a single numerical variable

          Key points:

          Shows frequency using bins
          Helps understand skewness

   6. KDE Plot (Density Curve)

          Use: Smooth distribution curve

          Key points:

          Often combined with histogram
          Shows probability density

   7. Box Plot

        Use: Distribution + outliers

        Shows:

        Median
        Quartiles (Q1, Q3)
        Outliers

        Key points:

        Best for comparing categories
        Easy outlier detection

   8. Violin Plot

        Use: Distribution + density

        Key points:

        Combines boxplot + KDE
        Shows shape of data distribution

   9. Pair Plot (VERY IMPORTANT)

        Use: Relationship between multiple features

        Key points:

        Auto plots scatter + histograms
        Great for quick EDA

   10. Heatmap

        Use: Correlation matrix

        Key points:

        Values range from -1 to +1
        Helps in feature selection
        Detects multicollinearity

   11. Regression Plot

        Use: Relationship + best fit line

        Key points:

        Shows trend + prediction line
        Useful in ML preprocessing

   12. Strip Plot

        Use: Distribution of points

        Key points:

        Shows individual data points
        Can overlap

   13. Swarm Plot

        Use: Better version of strip plot

        Key points:

        Avoids overlapping
        Cleaner visualization

   14. Pie Chart

        Use: Proportions

        Key points:

        Shows percentage distribution
        Not ideal for many categories
