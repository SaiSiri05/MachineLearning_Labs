```python
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import Isomap
import matplotlib.pyplot as plt

# Generate dataset
X, color = make_swiss_roll(n_samples=1000, noise=0.05)

# Apply Isomap
isomap = Isomap(n_neighbors=10, n_components=2)
X_reduced = isomap.fit_transform(X)

# Plot
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=color, cmap='viridis')
plt.title("Isomap Result")
plt.show()
```

```python
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import Isomap
import matplotlib.pyplot as plt

# More noise
X, color = make_swiss_roll(n_samples=1000, noise=0.3)

isomap = Isomap(n_neighbors=10, n_components=2)
X_reduced = isomap.fit_transform(X)

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=color, cmap='viridis')
plt.title("Isomap with High Noise")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=color, cmap='viridis')
plt.title("PCA Result")
plt.show()
```
